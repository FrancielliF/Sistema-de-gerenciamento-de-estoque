```python
class Produto:
  def __init__(self, nome, categoria, quantidade, preco, localizacao):
    self.nome = nome
    self.categoria = categoria
    self.quantidade = quantidade
    self.preco = preco
    self.localizacao = localizacao

class Estoque:
  def __init__(self):
    self.produtos = []

  def adicionar_produto(self, produto):
    self.produtos.append(produto)

  def atualizar_estoque(self, nome, nova_quantidade):
    for produto in self.produtos:
      if produto.nome == nome:
        produto.quantidade = nova_quantidade
        break

  def consultar_estoque(self, nome):
    for produto in self.produtos:
      if produto.nome == nome:
        return produto
    return None

  def listar_produtos(self):
    for produto in self.produtos:
      print

class Mivimentacao:
  def __init__(self):
    self.registros = []

  def registrar_movimentacao(self, movimentacao):
    self.registros.append(movimentacao)

def main():
  estoque = Estoque()
  movimentacao = Mivimentacao()

  while True:
    print("\nMenu:")
    print("1. Adicionar produto")
    print("2. Atualizar estoque")
    print("3. Consultar estoque")
    print("4. Listar produtos")
    print("5. Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
      nome = input("Nome do produto: ")
      categoria = input("Categoria do produto: ")
      quantidade = int(input("Quantidade em estoque: "))
      preco = float(input("Preço do produto: "))
      localizacao = input("Localização do produto: ")
      produto = Produto(nome, categoria, quantidade, preco, localizacao)
      estoque.adicionar_produto(produto)
      print("Produto adicionado com sucesso!")

    elif opcao == "2":
      nome = input("Nome do produto: ")
      nova_quantidade = int(input("Nova quantidade em estoque: "))
      produto = estoque.consultar_estoque(nome)
      if produto:
        estoque.atualizar_estoque(nome, nova_quantidade)
        print("Estoque atualizado com sucesso!")
      else:
        print("Produto não encontrado no estoque.")

    elif opcao == "3":
      nome = input("Nome do produto: ")
      produto = estoque.consultar_estoque(nome)
      if produto:
        print(f"Nome: {produto.nome}")
        print(f"Categoria: {produto.categoria}")
        print(f"Quantidade: {produto.quantidade}")
        print(f"Preço: {produto.preco}")
        print(f"Localização: {produto.localizacao}")
      else:
        print("Produto não encontrado no estoque.")

    elif opcao == "4":
      estoque.listar_produtos()
      print("Lista de produtos:")
      for produto in estoque.produtos:
        print(f"Nome: {produto.nome}")


    elif opcao == "5":
      print("Saindo do programa.")
      break

    else:
      print("Opção inválida. Tente novamente.")


if __name__ == "__main__":
  main()
```
